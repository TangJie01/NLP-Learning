```python
#txt文件读写

import os

# --- 环境自适应 ---
if os.path.exists('/kaggle/working'):
    WORK_DIR = '/kaggle/working/'
else:
    WORK_DIR = './'

# --- 写入 TXT 文件 ---
text = "This is my first test.\nThis is next line.\nThis is last line."
file_path = os.path.join(WORK_DIR, 'my_file.txt')

with open(file_path, 'w', encoding='utf-8') as f:
    f.write(text)

print(f"✅ TXT 文件已写入：{file_path}")

# --- 读取 TXT 文件 ---
with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()

print("\n--- TXT 文件内容 ---")
print(content)
```

```python
#csv文件读写

import os
import csv

# --- 环境自适应 ---
if os.path.exists('/kaggle/working'):
    WORK_DIR = '/kaggle/working/'
else:
    WORK_DIR = './'

# --- 准备数据 ---
header = ['姓名', '年龄', '性别']
data = [
    ['张三', 28, '男'],
    ['李四', 16, '女'],
    ['王五', 35, '男']
]

file_path = os.path.join(WORK_DIR, 'passengers.csv')

# --- 写入 CSV 文件 ---
with open(file_path, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)

print(f"✅ CSV 文件已写入：{file_path}")

# --- 读取 CSV 文件（方法一：csv 模块）---
print("\n--- 用 csv 模块读取 ---")
with open(file_path, 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# --- 读取 CSV 文件（方法二：pandas，推荐）---
import pandas as pd
print("\n--- 用 pandas 读取 ---")
df = pd.read_csv(file_path)
print(df)
```

```python
#文件追加，向文件中追加新内容，而不会删除或覆盖文件中已有的内容
#'a'模式，追加；'w'模式，清空重写

import os

# --- 环境自适应 ---
if os.path.exists('/kaggle/working'):
    WORK_DIR = '/kaggle/working/'
else:
    WORK_DIR = './'

file_path = os.path.join(WORK_DIR, 'my_file.txt')

# --- 追加内容 ---
with open(file_path, 'a', encoding='utf-8') as f:
    f.write("\n这是新追加的一行。\n")

print(f"✅ 内容已追加到: {file_path}")

# --- 读取并打印当前文件全部内容，验证是否成功 ---
print("\n--- my_file.txt 当前全部内容 ---")
with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()
    print(content)
```

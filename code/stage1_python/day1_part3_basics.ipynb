```python
#python-basic-def

def function():
    print('This is a function')
    a = 1 + 2
    print(a)

function()

def func(a,b):
    c = a - b
    print("c = ",c)

func(5,2)
func(b=1,a=2)


def fun():
    a=10
    print(a)
    return a+100

print(fun())
```

```python
def sale_car1(price,color,brand,is_second_hand):
    print('price:',price,
          'color:',color,
          'brand:',brand,
          "is_second_hand:",is_second_hand)

sale_car1(1000,'blue','BMW',True)


def sale_car2(price,color='red',brand='BMW',is_second_hand=False):
    print('price:',price,
          'color:',color,
          'brand:',brand,
          "is_second_hand:",is_second_hand)

sale_car2(2000)
sale_car2(2000,'blue')

#参数内已经定义的需要置后
def sale_car3(price,brand,color='red',is_second_hand=True):
    print('price:',price,
          'color:',color,
          'brand:',brand,
          "is_second_hand:",is_second_hand)

sale_car3(3000,'BMW')
```

```python
#全局变量

A = 100

def func():
    A = 20
    print("In the func, A is",A)
    return A+1000

print('Before the func, A is',A)
print(func())
print('After the func, A is',A)

def fun():
    global A
    A = 20
    print("In the func, A is",A)
    return A+1000

print('Before the fun, A is',A)
print(fun())
print('After the fun, A is',A)
```

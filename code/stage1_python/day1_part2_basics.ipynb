```python
#python-basic-while

condition = 1
while condition < 10 :
    print(condition);
    condition += 1
```

```python
#python-basic-for
#技巧：调整多行空格距离，ctrl+[向左，ctrl+]向右

list = [1,2,3,4,5]
for i in list:
    print(i)
    print(i,"in the list")

print("out")

#range(5)----[0,1,2,3,4]
for j in range(5):
    print(j)

#range(1,5)----[1,2,3,4]
for j in range(1,5):
    print(j)

#range(1,5,2)---距离/步长2，列举1，3
for j in range(1,5,2):
    print(j)
```

```python
#python-basic-if

x = 1
y = 2
z = 3

if x<y:
    print("x<y")

if x!=y:
    print("x!=y")

if x<y<z:
    print("x<y<z")

if x<z>y:
    print("x<z>y")
```

```python
#python-basic-if else
x = 1
y = 2
z = 3

if x==y:
    print("x==y")
else:
    print("x!=y")

#elif----else if
if x>y:
    print("x>y")
elif x>z:
    print("x>z")
else:
    print("x is the smallest number")

#elif,顺序最先匹配
if x>2:
    print("x>2")
elif x<2:
    print("x<2")
elif x==1:
    print("x=1")
else:
    print("no")

print("finish")
```

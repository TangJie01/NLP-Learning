```python
# 练习一：单个乘客信息录入（类 + 文件写入 + input）
import os

# ---------- 环境自适应 ----------
if os.path.exists('/kaggle/working'):
    WORK_DIR = '/kaggle/working/'
else:
    WORK_DIR = './'

# ---------- 定义乘客类 ----------
class Passenger:
    def __init__(self, name, age, gender, fare):
        # 初始化方法，将传入的参数绑定到实例上
        self.name = name
        self.age = age
        self.gender = gender
        self.fare = fare

    def info(self):
        # 返回格式化的字符串，方便写入文件
        return f"{self.name},{self.age}岁,{self.gender},票价{self.fare}元"


# ---------- 输入乘客信息 ----------
print("=== 请输入乘客信息 ===")
name = input("姓名：")
age = input("年龄：")
gender = input("性别：")
fare = input("票价：")

# ---------- 创建实例 ----------
p = Passenger(name, age, gender, fare)

# ---------- 追加写入文件 ----------
file_path = os.path.join(WORK_DIR, 'passenger_log.txt')
with open(file_path, 'a', encoding='utf-8') as f:
    f.write(p.info() + "\n")   # 每写入一条记录后换行

print(f"✅ 乘客信息已追加到：{file_path}")

# ---------- 读取并显示文件全部内容 ----------
print("\n--- passenger_log.txt 当前全部内容 ---")
with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()
    print(content)
```

```python
#练习二：简单计算器升级版（类 + 循环 + input）

import os

# ---------- 环境自适应 ----------
if os.path.exists('/kaggle/working'):
    WORK_DIR = '/kaggle/working/'
else:
    WORK_DIR = './'

# ---------- 定义计算器类 ----------
class Calculator:
    def add(self, x, y):
        return x + y

    def sub(self, x, y):
        return x - y

    def multi(self, x, y):
        return x * y

    def div(self, x, y):
        # 避免除零错误
        if y == 0:
            return "错误：除数不能为0"
        return x / y


# ---------- 主程序：交互式计算 ----------
calc = Calculator()                     # 创建计算器实例
history_path = os.path.join(WORK_DIR, 'calc_history.txt')

while True:
    print("\n=== 简单计算器 ===")
    x = input("请输入第一个数字（输入 exit 退出）：")
    if x.lower() == 'exit':             # 输入 exit 退出循环
        print("👋 程序已退出。")
        break

    op = input("请输入运算符（+、-、*、/）：")
    y = input("请输入第二个数字：")

    # 将输入转换为浮点数（如果转换失败则提示并重新开始）
    try:
        x_num = float(x)
        y_num = float(y)
    except ValueError:
        print("⚠️ 输入的不是有效数字，请重新输入。")
        continue

    # 根据运算符选择对应的计算方法
    if op == '+':
        result = calc.add(x_num, y_num)
    elif op == '-':
        result = calc.sub(x_num, y_num)
    elif op == '*':
        result = calc.multi(x_num, y_num)
    elif op == '/':
        result = calc.div(x_num, y_num)
    else:
        print("⚠️ 运算符输入错误，只支持 +、-、*、/")
        continue

    # 显示结果
    print(f"结果：{result}")

    # 构造历史记录字符串，并追加写入文件
    record = f"{x_num} {op} {y_num} = {result}"
    with open(history_path, 'a', encoding='utf-8') as f:
        f.write(record + "\n")
    print(f"📝 计算记录已保存到：{history_path}")
```

```python
#python-basic-class

class Calculator:
    name="Good Calculator"
    price=18
    
    def add(self,x,y):#不加self会报错
        print(self.name)
        result=x+y
        print(result)
        
    def sub(self,x,y):
        print(x-y)
        
    def multi(self,x,y):
        print(x*y)
        
    def div(self,x,y):
        print(x/y)

calcu=Calculator()
print(calcu.name)
print(calcu.price)
calcu.add(10,2)
calcu.sub(10,2)
calcu.multi(10,2)
calcu.div(10,2)
```

```python
#python-basic-init

class Calculator:
    name="Good Calculator"
    price=18
    
    def __init__(self,name,price,hight,width,weight):
        
        self.name=name
        self.price=price
        self.hi=hight
        self.wi=width
        self.we=weight
        self.add(1,2)#初始化时先执行函数，输出结果3

    def add(self,x,y):
        print(x+y)
        
c=Calculator('bad',12,13,14,15)
print(c.name)
print(c.price)
print(c.hi)
print(c.wi)
print(c.we)
```

```python
#python-basic-init

class Calculator:
    
    def __init__(self,name,price,hight,width=3,weight=5):
        
        self.name=name
        self.price=price
        self.hi=hight
        self.wi=width
        self.we=weight
        self.add(1,2)#初始化时先执行函数，输出结果3

    def add(self,x,y):
        print(x+y)
        
c=Calculator('bad',12,13)
print(c.name)
print(c.price)
print(c.hi)
print(c.wi)
print(c.we)
```
